```python
from modfilegen import GlobalVariables
import sqlite3
from pathlib import Path
import os
```

```python
# Base paths derived from the repository root, not from the kernel cwd
data_dir = "/data"
output_dir_stics = os.path.join(data_dir, "output_stics")
temp_dir_stics = os.path.join(output_dir_stics, "temp")
# Create output directories if they don't exist
os.makedirs(output_dir_stics, exist_ok=True)
os.makedirs(temp_dir_stics  , exist_ok=True)


output_dir_celsius = os.path.join(data_dir, "output_celsius")
temp_dir_celsius = os.path.join(output_dir_celsius, "temp")
# Create output directories if they don't exist
os.makedirs(output_dir_celsius, exist_ok=True)
os.makedirs(temp_dir_celsius  , exist_ok=True)


output_dir_dssat = os.path.join(data_dir, "output_dssat")
temp_dir_dssat = os.path.join(output_dir_dssat, "temp")
# Create output directories if they don't exist
os.makedirs(output_dir_dssat, exist_ok=True)
os.makedirs(temp_dir_dssat  , exist_ok=True)


# Database paths
master_input_db = os.path.join(data_dir, "MasterInput.db")
models_dict_db = os.path.join(data_dir, "ModelsDictionaryArise.db")
ori_mi = os.path.join(data_dir, "ori_MasterInput.db")
celsius_f = os.path.join(data_dir, "CelsiusV3nov17_dataArise.db")

# Cultivar/plant parameters folder
cultivars_folder_stics = os.path.join(data_dir, "cultivars", "stics")
cultivars_folder_dssat = os.path.join(data_dir, "cultivars", "dssat")

# Configuration
n_threads = 4  # Number of parallel threads
n_parts = 1    # Number of data partitions
dt = 0        # choice to delete intermediate files 0 (no) or 1 (yes)

# Verify files exist
print("Checking files...")
print(f"Master Input DB exists: {os.path.exists(master_input_db)}")
print(f"Models Dict DB exists: {os.path.exists(models_dict_db)}")
print(f"Stics Cultivars folder exists: {os.path.exists(cultivars_folder_stics)}")
print(f"Dssat Cultivars folder exists: {os.path.exists(cultivars_folder_dssat)}")
```

```python
# Set global variables
GlobalVariables["dbMasterInput"] = str(master_input_db)
GlobalVariables["dbModelsDictionary"] = str(models_dict_db)

GlobalVariables["nthreads"] = n_threads
GlobalVariables["dt"] = dt
GlobalVariables["parts"] = n_parts
GlobalVariables["ori_MI"] = master_input_db
GlobalVariables["dbCelsius"] = celsius_f

print("Global Variables configured:")
for key, value in GlobalVariables.items():
    print(f"  {key}: {value}")
```

```python
from modfilegen.Converter.CelsiusConverter import celsiusconverter

GlobalVariables["directorypath"] = str(output_dir_celsius)
GlobalVariables["tempDir"] = str(temp_dir_celsius)

def test_celsiusconverter():
    celsiusconverter.main()
    return 0

if __name__ == "__main__":
    test_celsiusconverter()
    print("test_celsiusconverter")
```

```python
# Run the converter
from modfilegen.Converter.SticsConverter import sticsconverter
GlobalVariables["directorypath"] = str(output_dir_stics)
GlobalVariables["pltfolder"] = str(cultivars_folder_stics)
GlobalVariables["tempDir"] = str(temp_dir_stics)
print("Starting STICS conversion...")
print("=" * 60)

try:
    sticsconverter.main()
    print("\n" + "=" * 60)
    print("✅ STICS conversion completed successfully!")
except Exception as e:
    print("\n" + "=" * 60)
    print(f"❌ Error during conversion: {e}")
    import traceback
    traceback.print_exc()
```
